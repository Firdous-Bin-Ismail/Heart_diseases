```python
df.encode = pd.get_dummies(df, drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
col=['Age','RestingBP','Cholesterol','MaxHR','Oldpeak']
```

```python
df.encode[col]= scaler.fit_transform(df.encode[col])
```

```python
df.encode.head()
```

```python
df.encode.astype(int)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score, classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
X = df.encode.drop('HeartDisease', axis=1)
y= df.encode['HeartDisease']
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.20, random_state=42)
```

```python
models= {
    "LogisticRegression": LogisticRegression(),
    "svm": SVC(),
    "naive_bayes": GaussianNB(),
    "tree": DecisionTreeClassifier(),
    "Kneighbors":KNeighborsClassifier()

}
```

```python
result= []
```

```python
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    Accuracy = accuracy_score(y_test, y_pred)
    f1_scor = f1_score(y_test, y_pred)
    result.append({
        "Model" : name,
        "Accuracy": Accuracy,
        "F1_value": f1_scor
    })
```
